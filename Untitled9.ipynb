```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.preprocessing import StandardScaler
```

```python
iris=pd.read_csv("Iris.csv")
iris.head()
```

```python
#select 2 features for visualization
X=iris[['SepalLengthCm','SepalWidthCm']]
y=iris['Species']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42,stratify=y)
```

```python
# Normalize feature
scaler=StandardScaler()
X_train_of=scaler.fit_transform(X_train)
X_test_of=scaler.fit_transform(X_test)
```

```python
# Use KNN classifiers
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train_of,y_train)
```

```python
y_pred_of=knn.predict(X_test_of)
```

```python
accuracy_of=accuracy_score(y_pred_of,y_test)
print("Accuracy Score:", accuracy_of)
```

```python
# Experiment different K values
k_value=range(1,16)
accuracy=[]
for k in k_value:
    model=KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train_of,y_train)
    y_pred=model.predict(X_test_of)
    
```

```python
accuracy=accuracy_score(y_test,y_pred)
print("accuracy score:", accuracy)
```
